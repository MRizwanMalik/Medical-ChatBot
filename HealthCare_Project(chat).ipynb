```python
X = dataset.drop('prognosis', axis=1)
y = dataset['prognosis']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pandas as pd
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4200)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
label_encoder = LabelEncoder()
```

```python
X_train
```

# **random Forest Model**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
rf_classifier = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42)

# Cross-validate the model
cv_scores = cross_val_score(rf_classifier, X_train_scaled, y_train, cv=5)

# Print cross-validation scores
print("Cross-Validation Scores:", cv_scores)
print("Mean CV Score:", cv_scores.mean())

# Train the model on the entire training set
rf_classifier.fit(X_train_scaled, y_train)

# Evaluate the model on the test set
test_accuracy = rf_classifier.score(X_test_scaled, y_test)
print("Test Accuracy:", test_accuracy)
```

# **Decision Tree MOdel**

```python

from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier(max_depth=30, random_state=42)

# Cross-validate the model
cv_scores = cross_val_score(dt_classifier, X_train_scaled, y_train, cv=5)

# Print cross-validation scores
print("Cross-Validation Scores:", cv_scores)
print("Mean CV Score:", cv_scores.mean())

# Train the model on the entire training set
dt_classifier.fit(X_train_scaled, y_train)

# Evaluate the model on the test set
test_accuracy = dt_classifier.score(X_test_scaled, y_test)
print("Test Accuracy:", test_accuracy)
```

# **K-nearestNeighbour **

```python
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors=30)

# Cross-validate the model
cv_scores = cross_val_score(knn_classifier, X_train_scaled, y_train, cv=5)

# Print cross-validation scores
print("Cross-Validation Scores:", cv_scores)
print("Mean CV Score:", cv_scores.mean())

# Train the model on the entire training set
knn_classifier.fit(X_train_scaled, y_train)

# Evaluate the model on the test set
test_accuracy = knn_classifier.score(X_test_scaled, y_test)
print("Test Accuracy:", test_accuracy)
```

```python
from sklearn.svm import SVC

# Create an SVM classifier with RBF kernel
svm_classifier_rbf = SVC(kernel='rbf', random_state=2)

# Cross-validate the model
cv_scores_rbf = cross_val_score(svm_classifier_rbf, X_train_scaled, y_train, cv=5)

# Print cross-validation scores
print("Cross-Validation Scores (RBF kernel):", cv_scores_rbf)
print("Mean CV Score (RBF kernel):", cv_scores_rbf.mean())

# Train the model on the entire training set
svm_classifier_rbf.fit(X_train_scaled, y_train)

# Evaluate the model on the test set
test_accuracy_rbf = svm_classifier_rbf.score(X_test_scaled, y_test)
print("Test Accuracy (RBF kernel):", test_accuracy_rbf)
```
